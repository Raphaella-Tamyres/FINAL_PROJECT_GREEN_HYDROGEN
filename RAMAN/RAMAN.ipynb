```python
Dados_MoS2_1H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-1H.txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_2H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-2H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_3H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-3H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_4H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-4H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_5H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-5H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_10H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-10H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_15H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-15H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_20H = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-20H (2).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
Dados_MoS2_P = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-P.txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
```

```python
# Lista de DataFrames
dataframes_1 = [
    Dados_MoS2_1H,
    Dados_MoS2_2H,
    Dados_MoS2_3H,
    Dados_MoS2_4H,
    Dados_MoS2_5H,
    Dados_MoS2_10H,
    Dados_MoS2_15H,
    Dados_MoS2_20H,
    Dados_MoS2_P
]

def moving_average(data, window_size):
    return data.rolling(window=window_size).mean()

# Normalizar a primeira coluna de cada DataFrame
for df in dataframes_1:
    min_val = df.iloc[:, 1].min()
    max_val = df.iloc[:, 1].max()
    df.iloc[:, 1] = (df.iloc[:, 1] - min_val) / (max_val - min_val)
    df.iloc[:, 1] = moving_average(df.iloc[:, 1], window_size=5)
```

```python
Dados_MoS2_UP = pd.read_csv("C:\JupyterLab\TCC\RAMAN\MoS2_BM_UP\MoS2-UP (1).txt", sep='\t', engine='python', header=46, encoding='ISO-8859-1', decimal=',')
```

```python
min_val = Dados_MoS2_UP.iloc[:, 1].min()
max_val = Dados_MoS2_UP.iloc[:, 1].max()
Dados_MoS2_UP.iloc[:, 1] = (Dados_MoS2_UP.iloc[:, 1] - min_val) / (max_val - min_val)
Dados_MoS2_UP.iloc[:, 1] = moving_average(Dados_MoS2_UP.iloc[:, 1], window_size=5)
```

```python
# Representação gráfica

plt.rcParams['font.family'] = 'Verdana'
plt.figure(figsize=(25, 15))

# Suavizar os dados e plotar
def plot_dados_suavizados(dados, janela, ordem, label, cor):
    eixo_x = dados.iloc[0:300, 0]
    eixo_y = dados.iloc[0:300, 1]
    eixo_y_suavizado = savgol_filter(eixo_y, janela, ordem)  # Aplicando suavização
    plt.plot(eixo_x, eixo_y_suavizado, label=label, linewidth=5, linestyle='-', color=cor)

# Definir parâmetros de suavização: janela de 15 pontos e polinômio de ordem 2
janela = 15  # Deve ser um número ímpar
ordem = 3

# Plotar os dados suavizados
plot_dados_suavizados(Dados_MoS2_1H, janela, ordem, r'$MoS_2$-1H', 'mediumturquoise')
plot_dados_suavizados(Dados_MoS2_2H, janela, ordem, r'$MoS_2$-2H', 'lightsalmon')
plot_dados_suavizados(Dados_MoS2_3H, janela, ordem, r'$MoS_2$-3H', 'violet')
plot_dados_suavizados(Dados_MoS2_4H, janela, ordem, r'$MoS_2$-4H', 'indianred')
plot_dados_suavizados(Dados_MoS2_5H, janela, ordem, r'$MoS_2$-5H', 'lightgreen')
plot_dados_suavizados(Dados_MoS2_10H, janela, ordem, r'$MoS_2$-10H', 'lightcoral')
plot_dados_suavizados(Dados_MoS2_15H, janela, ordem, r'$MoS_2$-15H', 'lightpink')
plot_dados_suavizados(Dados_MoS2_20H, janela, ordem, r'$MoS_2$-20H', 'yellow')
plot_dados_suavizados(Dados_MoS2_P, janela, ordem, r'$MoS_2$-Pristine', 'gray')

# Configuração do gráfico
plt.xlabel(r'Raman shift ($cm^{-1}$)', size=50)
plt.ylabel('Intensidade normalizada (a. u.)', size=50)
plt.tick_params(labelsize=40)
plt.rcParams['legend.fontsize'] = 40
plt.legend()

# Encontrar máximos nos intervalos definidos
intervalos = [
    (350, 400),
    (400, 450)
]

maximos_encontrados = []
anotacoes_especificas = [r"$E_{2G}$", r"$A_{1G}$"]

for intervalo in intervalos:
    dados_no_intervalo = Dados_MoS2_P[(Dados_MoS2_P.iloc[:, 0] >= intervalo[0]) & (Dados_MoS2_P.iloc[:, 0] <= intervalo[1])]
    if not dados_no_intervalo.empty:
        idx_maximo = dados_no_intervalo.iloc[:, 1].idxmax()
        maximos_encontrados.append((Dados_MoS2_P.iloc[idx_maximo, 0], Dados_MoS2_P.iloc[idx_maximo, 1]))

for i, ponto in enumerate(maximos_encontrados):
    plt.axvline(x=ponto[0], color='gray', linestyle='--', linewidth=5)
    plt.annotate(anotacoes_especificas[i], xy=(ponto[0], ponto[1]),
                 xytext=(ponto[0]-2, 0.85), fontsize=40, color='black', rotation=0,
                 verticalalignment='bottom', horizontalalignment='right')
    plt.annotate(round(ponto[0]), xy=(ponto[0], ponto[1]-0.1),
                 xytext=(ponto[0]-1, 0.01),
                 fontsize=30, color='black', rotation=90, verticalalignment='bottom', horizontalalignment='right')

if len(maximos_encontrados) == 2:
    y_avg = max(maximos_encontrados[0][1], maximos_encontrados[1][1])
    plt.hlines(y_avg, xmin=min(maximos_encontrados[0][0], maximos_encontrados[1][0]),
               xmax=max(maximos_encontrados[0][0], maximos_encontrados[1][0]),
               color='black', linestyle='--', linewidth=5)
    distancia = abs(maximos_encontrados[0][0] - maximos_encontrados[1][0])
    plt.text((maximos_encontrados[0][0] + maximos_encontrados[1][0]) / 2-37, y_avg+0.015,
             f'{distancia:.0f} $cm^-¹$', fontsize=40, color='black', ha='center')

plt.ylim(0, 0.9)
plt.grid()
plt.savefig('MoS2-BM.png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
