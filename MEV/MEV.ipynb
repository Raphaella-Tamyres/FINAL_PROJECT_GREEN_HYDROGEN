```python
# Dados
diameter_silica_nm = 400  # nm
mass_silica_g = 0.15      # g (150 mg)
density_silica_g_cm3 = 2  # g/cm³

# Conversões
diameter_silica_cm = diameter_silica_nm * 1e-7  # Convertendo nm para cm
radius_silica_cm = diameter_silica_cm / 2       # Raio em cm

# Volume de uma nanopartícula de sílica (esfera)
volume_silica_particle_cm3 = (4/3) * math.pi * (radius_silica_cm ** 3)

# Massa de uma única nanopartícula de sílica (massa = volume * densidade)
mass_silica_particle_g = volume_silica_particle_cm3 * density_silica_g_cm3

# Número total de nanopartículas de sílica
num_particles = mass_silica_g / mass_silica_particle_g

# Área superficial de uma única nanopartícula de sílica (área = 4 * π * r^2)
surface_area_particle_cm2 = 4 * math.pi * (radius_silica_cm ** 2)

# Área superficial total das nanopartículas de sílica
total_surface_area_silica_cm2 = num_particles * surface_area_particle_cm2

# Assumindo uma densidade superficial aproximada para uma monocamada de MoS2 (em g/cm²).
# Referências sugerem cerca de 3.2e-6 g/cm² para MoS2
surface_density_mos2_g_cm2 = 3.2e-6  # g/cm²

# Massa necessária de MoS2 para recobrir a área total das nanopartículas de sílica
mass_mos2_needed_g = total_surface_area_silica_cm2 * surface_density_mos2_g_cm2

# Convertendo para mg
mass_mos2_needed_mg = mass_mos2_needed_g * 1000  # g para mg
print(r'Massa de MoS2 para recobrir a sílica', round(mass_mos2_needed_mg,0))
```
