```python
# Definindo os dados para cada gráfico
x_1 = Dados_S1.iloc[40:52, 4]
y_1 = Dados_S1.iloc[40:52, 5]
x_5 = Dados_S5.iloc[41:53, 4]
y_5 = Dados_S5.iloc[41:53, 5]
x_7 = Dados_S7.iloc[41:53, 4]
y_7 = Dados_S7.iloc[41:53, 5]
x_8 = Dados_S8.iloc[43:52, 4]
y_8 = Dados_S8.iloc[43:52, 5]

# Função para suavizar e encontrar picos
def process_data(x, y):
    x_smooth = np.linspace(x.min(), x.max(), 100)
    spline = make_interp_spline(x, y)
    y_smooth = spline(x_smooth)
    
    # Encontrar picos
    peaks, _ = find_peaks(y_smooth)
    peak_x = x_smooth[peaks]
    peak_y = y_smooth[peaks]
    
    return x_smooth, y_smooth, peak_x, peak_y

# Processar os dados
x_1_smooth, y_1_smooth, peaks_1_x, peaks_1_y = process_data(x_1, y_1)
x_5_smooth, y_5_smooth, peaks_5_x, peaks_5_y = process_data(x_5, y_5)
x_7_smooth, y_7_smooth, peaks_7_x, peaks_7_y = process_data(x_7, y_7)
x_8_smooth, y_8_smooth, peaks_8_x, peaks_8_y = process_data(x_8, y_8)

# Plotar os dados
plt.rcParams['font.family'] = 'Verdana'
plt.figure(figsize=(15, 15))
plt.plot(x_1_smooth, y_1_smooth, label=r'$SiO_2$ (1)', linewidth=5, linestyle='-', color='mediumturquoise')
plt.plot(x_5_smooth, y_5_smooth, label=r'$SiO_2$ (2)', linewidth=5, linestyle='-', color='lightsalmon')
plt.plot(x_7_smooth, y_7_smooth, label=r'$SiO_2$ (3)', linewidth=5, linestyle='-', color='violet')
plt.plot(x_8_smooth, y_8_smooth, label=r'$SiO_2$ (4)', linewidth=5, linestyle='-', color='lightgreen')

# Adicionar os picos ao gráfico
for peaks_x, peaks_y in zip([peaks_1_x, peaks_5_x, peaks_8_x],
                             [peaks_1_y, peaks_5_y, peaks_8_y]):
    plt.scatter(peaks_x, peaks_y, color='black', s=100)  # Marcadores para os picos
    for (px, py) in zip(peaks_x, peaks_y):
        plt.annotate(f'{px:.0f} \n nm', (px, py), textcoords="offset points", xytext=(0,10), ha='center', fontsize=30, color='black', rotation=0)

# Configurações do gráfico
plt.xlabel('Tamanho (d.nm)', size=40)
plt.ylabel('Intensidade (%)', size=40)
plt.title('', size=40)
plt.tick_params(labelsize=30)
plt.rcParams['legend.fontsize'] = 30
plt.xlim(101,950)
plt.ylim(-1,27)
plt.grid()
plt.legend()
plt.savefig('DLS - sílica coloidal.png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```
