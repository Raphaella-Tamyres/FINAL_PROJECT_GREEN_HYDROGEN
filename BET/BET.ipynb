```python
# Função para separar a adsorção e dessorção com base no valor máximo de adsorção
def separate_adsorption_desorption(data):
    # Encontrar o valor máximo da quantidade adsorvida
    max_adsorbed = data.iloc[:, 2].max()

    # Encontrar o índice do valor máximo
    max_index = data.iloc[:, 2].idxmax()

    # Separar os dados em adsorção (antes do valor máximo) e dessorção (depois do valor máximo)
    adsorption_data = data[:max_index + 1]  # Até o ponto máximo
    desorption_data = data[max_index:]      # Após o ponto máximo
    
    return adsorption_data, desorption_data

# Separar os dados de adsorção e dessorção para cada amostra (substitua 'Pristine', 'MoS2_20h', etc.)
Pristine_adsorption, Pristine_desorption = separate_adsorption_desorption(Pristine)
MoS2_20h_adsorption, MoS2_20h_desorption = separate_adsorption_desorption(MoS2_20h)
MoS2_5h_adsorption, MoS2_5h_desorption = separate_adsorption_desorption(MoS2_5h)
MoS2_UP_adsorption, MoS2_UP_desorption = separate_adsorption_desorption(MoS2_UP)
MoS2_NS_A_adsorption, MoS2_NS_A_desorption = separate_adsorption_desorption(MoS2_NS_amonio)
MoS2_NS_S_adsorption, MoS2_NS_S_desorption = separate_adsorption_desorption(MoS2_NS_sodio)

# Configuração do gráfico
plt.rcParams['font.family'] = 'Verdana'
plt.figure(figsize=(25, 15))

# Plotar as curvas de adsorção normalizadas no mesmo gráfico
plt.plot(Pristine_adsorption.iloc[:, 0], Pristine_adsorption.iloc[:, 2], label=r'$MoS_2$-Pristine (Adsorção)', 
         linewidth=3, linestyle='-', marker='o', color='mediumturquoise')
plt.plot(MoS2_20h_adsorption.iloc[:, 0], MoS2_20h_adsorption.iloc[:, 2], label=r'$MoS_2$(1)-20H (Adsorção)', 
         linewidth=3, linestyle='-', marker='s', color='hotpink')
plt.plot(MoS2_5h_adsorption.iloc[:, 0], MoS2_5h_adsorption.iloc[:, 2], label=r'$MoS_2$(2)-5H (Adsorção)', 
         linewidth=3, linestyle='-', marker='^', color='violet')
plt.plot(MoS2_UP_adsorption.iloc[:, 0], MoS2_UP_adsorption.iloc[:, 2], label=r'$MoS_2$-UP (Adsorção)', 
         linewidth=3, linestyle='-', marker='D', color='indianred')
plt.plot(MoS2_NS_A_adsorption.iloc[:, 0], MoS2_NS_A_adsorption.iloc[:, 2], label=r'$MoS_2$-NS amônio (Adsorção)', 
         linewidth=3, linestyle='-', marker='p', color='lightgreen')
plt.plot(MoS2_NS_S_adsorption.iloc[:, 0], MoS2_NS_S_adsorption.iloc[:, 2], label=r'$MoS_2$-NS sódio (Adsorção)', 
         linewidth=3, linestyle='-', marker='p', color='orange')

# Plotar as curvas de dessorção no mesmo gráfico
plt.plot(Pristine_desorption.iloc[:, 0], Pristine_desorption.iloc[:, 2], label=r'$MoS_2$-Pristine (Dessorção)', 
         linewidth=3, linestyle='--', marker='o', color='mediumturquoise')
plt.plot(MoS2_20h_desorption.iloc[:, 0], MoS2_20h_desorption.iloc[:, 2], label=r'$MoS_2$(1)-20H (Dessorção)', 
         linewidth=3, linestyle='--', marker='s', color='hotpink')
plt.plot(MoS2_5h_desorption.iloc[:, 0], MoS2_5h_desorption.iloc[:, 2], label=r'$MoS_2$(2)-5H (Dessorção)', 
         linewidth=3, linestyle='--', marker='^', color='violet')
plt.plot(MoS2_UP_desorption.iloc[:, 0], MoS2_UP_desorption.iloc[:, 2], label=r'$MoS_2$-UP (Dessorção)', 
         linewidth=3, linestyle='--', marker='D', color='indianred')
plt.plot(MoS2_NS_A_desorption.iloc[:, 0], MoS2_NS_A_desorption.iloc[:, 2], label=r'$MoS_2$-NS amônio (Dessorção)', 
         linewidth=3, linestyle='--', marker='p', color='lightgreen')
plt.plot(MoS2_NS_S_desorption.iloc[:, 0], MoS2_NS_S_desorption.iloc[:, 2], label=r'$MoS_2$-NS sódio (Dessorção)', 
         linewidth=3, linestyle='--', marker='p', color='orange')

# Configurações do gráfico
plt.xlabel("Pressão Relativa (P/Po)", size=40)
plt.ylabel("Quantidade Adsorvida (cm³/g STP)", size=40)
plt.title('', size=40)
plt.tick_params(labelsize=30)
plt.rcParams['legend.fontsize'] = 30
plt.legend(ncol=2)

# Salvar e exibir a figura
plt.savefig('Curva de adsorção e dessorção - BET.png', dpi=300, bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
# Dados
amostras = [
    r"$MoS_2$ - Pristine",
    r"$MoS_2$ (1) - 20H",
    r"$MoS_2$ (2) - 5H",
    r"$MoS_2$ - UP",
    r"$MoS_2$ - NS (amônio)",
    r"$MoS_2$ - NS (sódio)"
]

area_superficie = [5.1610, 29.8680, 16.1367, 6.9477, 43.2744, 108.5982]  # m²/g
volume_poro = [0.024447, 0.150919, 0.069468, 0.034768, 0.116157, 0.239881]  # cm³/g
tamanho_poro = [322.689, 186.979, 185.118, 256.359, 93.455, 71.509]  # Å

# Ordenar os dados com base nos valores (os índices ordenados)
sorted_indices_area = np.argsort(area_superficie)
sorted_indices_volume = np.argsort(volume_poro)
sorted_indices_tamanho = np.argsort(tamanho_poro)

# Dados ordenados
amostras_sorted_area = np.array(amostras)[sorted_indices_area]
area_superficie_sorted = np.array(area_superficie)[sorted_indices_area]
volume_poro_sorted = np.array(volume_poro)[sorted_indices_volume]
tamanho_poro_sorted = np.array(tamanho_poro)[sorted_indices_tamanho]

# Configuração do gráfico
x_area = np.arange(len(amostras_sorted_area))
x_volume = np.arange(len(amostras_sorted_area))
x_tamanho = np.arange(len(amostras_sorted_area))
largura = 0.35  # largura das barras

# Criar a figura e os subgráficos (axes)
fig, axs = plt.subplots(1, 3, figsize=(25, 8))

# Gráfico 1 - Área de Superfície
axs[0].bar(x_area, area_superficie_sorted, largura, color='violet')
axs[0].set_title('Área de Superfície', fontsize=30)
axs[0].set_xticks(x_area)
axs[0].set_xticklabels(amostras_sorted_area, rotation=30, ha='right', fontsize=20)
axs[0].set_ylabel('m²/g', fontsize=20)
axs[0].tick_params(axis='both', labelsize=20)

# Adicionar os valores nas barras com ajuste na posição
for barra in axs[0].patches:
    altura = barra.get_height()
    axs[0].text(barra.get_x() + barra.get_width()/2, altura + (altura * 0.05), f'{altura:.2f}', ha='center', va='bottom', fontsize=16)
# Ajuste dos limites
axs[0].set_ylim(0, max(area_superficie_sorted) * 1.2)

# Gráfico 2 - Volume de Poro
axs[1].bar(x_volume, volume_poro_sorted, largura, color='mediumturquoise')
axs[1].set_title('Volume de Poro', fontsize=30)
axs[1].set_xticks(x_volume)
axs[1].set_xticklabels(amostras_sorted_area, rotation=30, ha='right', fontsize=20)
axs[1].set_ylabel('cm³/g', fontsize=20)
axs[1].tick_params(axis='both', labelsize=20)

# Adicionar os valores nas barras com ajuste na posição
for barra in axs[1].patches:
    altura = barra.get_height()
    axs[1].text(barra.get_x() + barra.get_width()/2, altura + (altura * 0.05), f'{altura:.2f}', ha='center', va='bottom', fontsize=16)
# Ajuste dos limites
axs[1].set_ylim(0, max(volume_poro_sorted) * 1.2)

# Gráfico 3 - Tamanho de Poro
axs[2].bar(x_tamanho, tamanho_poro_sorted, largura, color='lightsalmon')
axs[2].set_title('Tamanho de Poro', fontsize=30)
axs[2].set_xticks(x_tamanho)
axs[2].set_xticklabels(amostras_sorted_area, rotation=30, ha='right', fontsize=20)
axs[2].set_ylabel('Å', fontsize=20)
axs[2].tick_params(axis='both', labelsize=20) 

# Adicionar os valores nas barras com ajuste na posição
for barra in axs[2].patches:
    altura = barra.get_height()
    axs[2].text(barra.get_x() + barra.get_width()/2, altura + (altura * 0.05), f'{altura:.2f}', ha='center', va='bottom', fontsize=16)
# Ajuste dos limites
axs[2].set_ylim(0, max(tamanho_poro_sorted) * 1.2)

# Ajuste da figura e exibição
plt.tight_layout()

# Salvar o gráfico ajustado
plt.savefig('BET área, volume e poro.png', bbox_inches='tight')
plt.show()
```
